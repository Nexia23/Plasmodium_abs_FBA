```python
import matplotlib.pyplot as plt
import matplotlib.offsetbox as of

#plt.style.use('fivethirtyeight')

# Set-up in a similar way to the LaTeX example
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = 'Ubuntu'
plt.rcParams['font.sans-serif'] = []
plt.rcParams['font.monospace'] = 'Ubuntu Mono'
plt.rcParams['axes.labelsize'] = 10
plt.rcParams['font.size'] = 10
plt.rcParams['legend.fontsize'] = 8
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['xtick.labelsize'] = 8
plt.rcParams['ytick.labelsize'] = 8
plt.rcParams['figure.titlesize'] = 12

fig = plt.figure(figsize=(8, 6), dpi=400)
plt.bar([1, 2, 3, 4], [125, 100, 90, 110], label="Product A",
        width=0.5, align='center')
ax1 = plt.gca()

# Split each line of text into a list of strings and formats
cmt_line1_txt = ["Our '", "Green Shoots", "' Marketing campaign, started in ",]
cmt_line1_fmt = [{'ha': 'left', 'va': 'bottom'},
            {'color':'#0BFF01', 'weight':'bold', 'ha':'left', 'va':'bottom'},
            {'color':'black', 'ha':'left', 'va':'bottom'},
            ]
cmt_line2_txt = ["Q3 ", "showed some positive impact in ",
            "Q4", ". Further ", "positive ",
            ]
cmt_line2_fmt = [
    {'color':'#BBBBBB', 'ha':'left', 'va':'bottom'},
    {'color':'black', 'ha':'left', 'va':'bottom'},
    {'color':'#BBBBBB', 'ha':'left', 'va':'bottom'},
    {'color':'black', 'ha':'left', 'va':'bottom'},
    {'color':'#011EFE', 'fontweight': 'bold', 'ha':'left', 'va':'bottom'},
]

cmt_line3_txt = ["impact", " is expected in ", "later quarters."]
cmt_line3_fmt=[
{'color':'blue', 'fontweight': 'bold', 'ha':'left', 'va':'bottom'},
{'color':'black', 'ha':'left', 'va':'bottom'},
{'style': 'italic', 'ha':'left', 'va':'bottom'}
]

# Create a single list for each line of text and format
# The list for each line is a number of TextArea objects
cmt_line1_lst = []
for txt, frmt in zip(cmt_line1_txt, cmt_line1_fmt):
    print(txt+str(frmt))
    cmt_line1_lst.append(of.TextArea(txt, textprops=dict(frmt)))
print(cmt_line1_lst)
cmt_line2_lst = []
for txt, frmt in zip(cmt_line2_txt, cmt_line2_fmt):
    cmt_line2_lst.append(of.TextArea(txt, textprops=dict(frmt)))

cmt_line3_lst = []
for txt, frmt in zip(cmt_line3_txt, cmt_line3_fmt):
    cmt_line3_lst.append(of.TextArea(txt, textprops=dict(frmt)))

# Pack each Text object in a line into a Horizontal Packer
texts_hbox1 = of.HPacker(children=cmt_line1_lst, pad=0, sep=0)
texts_hbox2 = of.HPacker(children=cmt_line2_lst, pad=0, sep=0)
texts_hbox3 = of.HPacker(children=cmt_line3_lst, pad=0, sep=0)

# Put the Horizontally Packed lines into a list
texts_line_store = []
texts_line_store.append(texts_hbox1)
texts_line_store.append(texts_hbox2)
texts_line_store.append(texts_hbox3)

# Put the lines of text into a Vertical Box to get the final result
texts_vbox = of.VPacker(children=texts_line_store, pad=2, sep=0)

# Put the final box on the figure
annot2_bbox=dict(facecolor='yellow', boxstyle='round', edgecolor='0.5', alpha=1)
arrow2=dict(arrowstyle='-|>, head_width=0.3', linewidth=1, facecolor='black')
annot2 = of.AnnotationBbox(texts_vbox, xy=(4, 80),xybox=(2.50, 120),
                        xycoords='data', boxcoords='data',
                        arrowprops=arrow2, bboxprops=annot2_bbox)
ax1.add_artist(annot2)

# Standard description of the plot
# Set xticks, Font is set globally
plt.xticks([1, 2, 3, 4], ['Q1', 'Q2', 'Q3', 'Q4'])
plt.xlabel('Time - FY quarters')
plt.ylabel('Sales - unadjusted')
plt.title('Total sales by quarter')
plt.legend(loc='best')
```

```python
def index_containing_substring(the_list, substring):
    for i, s in enumerate(the_list):
        if substring in s:
              return i
    return -1
```

```python
from itertools import chain 
import itertools as it

t2 = t.iloc[:,133:151]
col_total = [list(g) for k, g in it.groupby(t2.columns, lambda x: 'total' in x) if k]
col_total_k=np.array(col_total).flatten()
col_vals = [list(g) for k, g in it.groupby(t2.columns, lambda x: 'total' in x) if not k]
col_vals_fl=list(chain.from_iterable(col_vals))

labels_lips1=[]
plots=[]
for i,name in enumerate(names):
    index = index_containing_substring(col_total_k, name)
    
    if index != -1:
        es = col_total[index]
        df = t2[es + col_vals[index]]
        df = df.mean().T.sort_values(ascending=False).T
        
        labels_lips1.append(df.index[1:].to_list())
        values = df[1:]*t3.mean().values[i]/100
        print(df[col_vals[index]])
        plots.append(values)
    else:
        print(name)
        plots.append([t3.mean().values[i]])
        labels_lips1.append(['\% '+name])
        
flatten_list = list(chain.from_iterable(plots))
print(flatten_list)
labels_lips = list(chain.from_iterable(labels_lips1))
labels_lips1
```

```python
size = 0.1
r=2
# Data to plot
labels = t3.columns.to_list()
sizes = t3.mean().values
labels_gender = labels_lips
sizes_gender = flatten_list
colors = ['#e60000', '#e67300', '#00cc00', '#005ce6']
colors_gender = ['#ff4d4d','#ff3333', '#ff1a1a', '#ffbf80','#ffb366', '#ffa64d','#ff9933','#99ff99','#3399ff','#1a8cff','#0080ff']
# Plot
fig, ax =  plt.subplots(figsize=(7, 7), subplot_kw=dict(aspect="equal"))



plt.pie(sizes_gender,colors = colors_gender, startangle=0, radius=r, shadow=True,
        textprops={'size': 16})

wedges, texts = ax.pie( np.array(sizes),  colors = colors, radius=r*0.7, startangle=0, shadow=True)

#plt.pie(np.array(sizes), colors = 'black', radius=r*0.6, startangle=0)

bbox_props = dict(boxstyle="round", fc="black", ec="k", lw=0.2, alpha=0.7)

kw = dict(arrowprops=dict(arrowstyle="-",
                            color="0",
                            patchB=None,
                            shrinkB=0),
          bbox=bbox_props, zorder=1, va="center")
it = 0
for i, p in enumerate(wedges):
    
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=1,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    #kw['bbox']['fc']='#ffffe6'
    
    texts_line_store = []
    
    for ind,nme in enumerate([labels[i]]+labels_lips1[i]):
        if ind == 0:
            percent = sizes[i]/sizes.sum()*100
            txt = str(nme)+' '+ str(np.round(percent,2))+'\%' 
            frmt = {'color':colors[i], 'fontweight': 'bold', 'ha':'left', 'va':'bottom','fontsize':29}
            cmt_line3_lst = [of.TextArea(txt, textprops=dict(frmt))]
            
            texts_hbox3 = of.HPacker(children=cmt_line3_lst, pad=0, sep=0)
            
            texts_line_store.append(texts_hbox3)
        else:
            percent = np.array(plots[i][ind-1])/np.array(plots[i]).sum()*100
            txt = str(np.round(percent,2))+str(nme)
            frmt = {'color':colors_gender[it], 'fontweight': 'bold', 'ha':'left', 'va':'bottom', 'fontsize':25}
            it+=1
            cmt_line3_lst = [of.TextArea(txt, textprops=dict(frmt))]
            
            texts_hbox3 = of.HPacker(children=cmt_line3_lst, pad=0, sep=0)
            
            texts_line_store.append(texts_hbox3)
        
        texts_vbox = of.VPacker(children=texts_line_store, pad=2, sep=0)
        
        # Put the final box on the figure
        #annot2_bbox=dict(facecolor='yellow', boxstyle='round', edgecolor='0.5', alpha=1)
        #arrow2=dict(arrowstyle='-|>, head_width=0.3', linewidth=1, facecolor='black')
        
        annot2 = of.AnnotationBbox(texts_vbox, xy=(np.sign(x), y),xybox=(np.sign(x), y),
                                xycoords='data', boxcoords=('data'),
                                arrowprops=kw["arrowprops"], bboxprops=bbox_props)
        

        ax.add_artist(annot2)
        #s = ax.annotate(text, color= colors[i], xy=(x, y), xytext=(1.8*r*np.sign(x), 1.6*r*y),size = 20,
                   # horizontalalignment=horizontalalignment, **kw)
        
#centre_circle = plt.Circle((0,0),1.24,color='white', fc='white',linewidth=2)
#fig.gca().add_artist(centre_circle)
print(it)
plt.show()
# save
from io import BytesIO
f = BytesIO()
plt.savefig(f, format="svg")
```
